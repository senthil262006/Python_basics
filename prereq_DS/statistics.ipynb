# Data:



Types of Data:


* Quantitative Data: 
 * Numerical data that can be measured. Example: height, weight, temperature.
 <br>
 <br>
* Qualitative Data: 
     * Categorical data that describes characteristics or qualities. Example: colors, types of fruit, yes/no responses.



* Scales of Measurement:
<br>

    * __Nominal:__ Categories without a specific order (e.g., types of fruits).

    * __Ordinal:__ Categories with a specific order, but without consistent differences between categories (e.g., rankings).

    * __Interval:__ Numerical data with consistent differences between values, but no true zero point (e.g., temperature in Celsius).

    * __Ratio:__ Numerical data with a true zero point, allowing for meaningful comparisons of magnitude (e.g., weight).

* <font size="3">Measures of Dispersion:</font>
<br>

    * __Range:__                 The difference between the highest and lowest values.
    * __Variance:__               The average of the squared differences from the mean.
    * __Standard Deviation (σ):__ The square root of the variance, indicating how spread out the data points are.
    


<font size="3"> __Interpretation:__</font>
<br>
<br>
<font size="2"> __Low Standard Deviation:__</font>
<br>

* Indicates that the data points are close to the mean.


 <font size="2"> __High Standard Deviation:__</font>
<br>
* Indicates that the data points are spread out over a wide range of values.

```python
def variance(data):
    mean = sum(data) / len(data)
    v = sum((i - mean)**2 for i in data) / len(data)
    return v

variance([1,2,3,4,5,23,7,8,9,9,0,5,3,2,5,6,8,5])
```

 <font size=3>  __five-number summary__ </font>
<br>
* The five-number summary is a quick and informative statistical summary of a dataset. 
<br>
* It consists of the following five values:
<br>
* __Minimum:__ The smallest data point.

* __First Quartile (Q1):__ The median of the lower half of the data (25th percentile).

* __Median (Q2):__ The middle value of the data set (50th percentile).

* __Third Quartile (Q3):__ The median of the upper half of the data (75th percentile).

* __Maximum:__ The largest data point.



```python
<font size=3>  __Steps to Calculate the Five-Number Summary__ </font>
<br>
* __Order the data:__ Arrange the data in ascending order.
* __Identify the Minimum and Maximum:__ These are simply the smallest and largest values in the data set.
* __Find the Median:__ If the number of observations is odd, the median is the middle value. If even, 
<br>
* __it's the average of the two middle values.__
<br>
* __Calculate Q1:__ The median of the lower half of the data (not including the median if the number of observations is odd).
* __Calculate Q3:__ The median of the upper half of the data(again, not including the median if the number of observations is odd).
```

__Let’s calculate the five-number summary and create a box plot using Python:__

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = [12, 15, 14, 10, 8, 11, 13, 16, 18, 14, 19]

# Five-number summary
minimum = np.min(data)
q1 = np.percentile(data, 25)
median = np.median(data)
q3 = np.percentile(data, 75)
maximum = np.max(data)

five_number_summary = (minimum, q1, median, q3, maximum)
print("Five-Number Summary:", '(minimum, q1, median, q3, maximum)')
print("Five-Number Summary:", five_number_summary)

# Box plot
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

# frequency distribution

<font size=3>  __Key Components of Frequency Distribution__ </font> 


* __Class Intervals (or Bins):__ For continuous data, the range of values is divided into intervals or bins.
* __Frequency:__ The number of data points that fall within each class interval or belong to each category.
* __Relative Frequency:__ The proportion of the total number of data points that fall within each interval or category, calculated as the frequency of the interval divided by the total number of observations.
* __Cumulative Frequency:__ The sum of frequencies for all class intervals up to and including the current interval.

<font size=3> __Steps to Create a Frequency Distribution:__ </font>
    


*1.* __Choose class intervals:__ For example, intervals might be 40-49, 50-59, 60-69, etc.

*2.* __Tally the data:__ Count how many data points fall into each interval.

*3.* __Calculate frequency:__ Record the number of occurrences for each interval.

*4.* __Calculate relative frequency (optional):__ Divide each frequency by the total number of data points.

*5.* __Calculate cumulative frequency (optional):__ Sum the frequencies up to each interval.

<font size= 3> __Here's how you might create a frequency distribution for the given data using Python:__</font>

```python
import pandas as pd

# Sample data
data = [45, 67, 45, 70, 80, 67, 80, 90, 95, 95, 100, 70, 55, 75]

# Create a DataFrame
df = pd.DataFrame(data, columns=['Scores'])

# Define bins (class intervals)
bins = [40, 50, 60, 70, 80, 90, 100]

# Create frequency distribution
frequency = pd.cut(df['Scores'], bins=bins, right=False).value_counts().sort_index()

# Convert to DataFrame for better presentation
freq_df = pd.DataFrame(frequency).reset_index()
freq_df.columns = ['Score Range', 'Frequency']

# Add Relative Frequency
freq_df['Relative Frequency'] = freq_df['Frequency'] / freq_df['Frequency'].sum()

# Add Cumulative Frequency
freq_df['Cumulative Frequency'] = freq_df['Frequency'].cumsum()
print(f"date :{data}")
print()
print(freq_df)
pd.cu
```
