```python
#step 1: Import the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
#read the data
dataset = pd.read_excel("Superstore_USA.xlsx")
```

```python
dataset
```

```python
dataset.head(10)
```

```python
#missing values
dataset.isnull().sum()
```

```python
dataset["Product Base Margin"].fillna(dataset["Product Base Margin"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.columns
```

```python
#based on the priority customer
dataset['Order Priority'].value_counts()
```

```python
dataset['Order Priority'].unique()
```

```python
dataset['Order Priority']=dataset['Order Priority'].replace('Critical ','Critical')
```

```python
dataset['Order Priority'].value_counts()
```

```python
#Graphical analysis
```

<h1>Order Priority</h1>

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Order Priority",data=dataset)
plt.title("Count of Order Priority")
plt.savefig("Order_Priority.png")
plt.show()
```

<h1>Ship Mode</h1>

```python
dataset["Ship Mode"].value_counts()
```

```python
x_axis=dataset["Ship Mode"].value_counts().index
```

```python
x_axis
```

```python
y_axis=dataset["Ship Mode"].value_counts().values
```

```python
y_axis
```

```python
plt.figure(figsize=(5,4))
plt.pie(y_axis,labels=x_axis,startangle=60,autopct="%0.2f%%")
plt.title("Percentage of ship Mode")
plt.legend(loc=2)
plt.savefig("Percentage of ship Mode.png")
plt.show()

```

```python
#Bivariate Analysis
sns.countplot(x="Ship Mode",data=dataset)
```

```python

sns.countplot(x="Ship Mode",data=dataset,hue="Product Category")
plt.title("Product Category wise ship mode")
plt.savefig("Product Category wise ship mode.png")
plt.show()
```

```python
#Customer Segment
sns.countplot(x="Customer Segment",data=dataset)
plt.show()
```

```python
#Customer Segment
sns.countplot(x="Customer Segment",data=dataset,hue="Product Category")
plt.show()
```

```python
#Product Category
plt.figure(figsize=(10,8))
sns.countplot(x="Product Category",data=dataset,hue="Product Sub-Category")
```
